```python
import numpy as np
import pandas as pd
```

```python

hospital_df = pd.read_csv('hospital-costs.csv')
```

```python
hospital_df.head()
```

```python
hospital_df["Total Charges"] = hospital_df["Discharges"] * hospital_df["Mean Charge"]
```

```python
hospital_df["Total Cost"] = hospital_df["Discharges"] * hospital_df["Mean Cost"]
```

```python
hospital_df["Markup"] = hospital_df["Total Charges"] / hospital_df["Total Cost"]
```

```python
hospital_df[hospital_df["Markup"] == hospital_df["Markup"].aggregate(min)]
```

```python
hospital_df[hospital_df["Markup"] == hospital_df["Markup"].aggregate(max)]
```

```python
description_groups = hospital_df.groupby("APR DRG Description")
```

```python
description_groups["Discharges"].sum().sort_values(ascending=False).head(10)
```

```python
net = pd.DataFrame(hospital_df[["Facility Name", "Total Charges", "Total Cost"]])
```

```python
group = net.groupby("Facility Name")
```

```python
money = group[["Total Cost", "Total Charges"]].sum()
```

```python
money["Net Income"] = money["Total Charges"] - money["Total Cost"]
```
